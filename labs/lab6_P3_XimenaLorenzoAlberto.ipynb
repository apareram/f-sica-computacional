```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cuadrados(x1, y1, L, n):
    '''
        x1, y1: coordenadas del vértice superior izquierdo
        L: longitud del lado del cuadrado a subdividir
        n: número de iteraciones
    '''
    if n > 0:
        x2 = x1 + L/2
        y2 = y1 
        x3 = x2 
        y3 = y2 + L/2

        cuadrados(x1, y1, L/2, n-1)
        cuadrados(x2, y2, L/2, n-1)
        cuadrados(x3, y3, L/2, n-1)

    else:
        cuadrado = plt.Rectangle((x1, y1), L, L, color='r', fill=True)
        ax.add_patch(cuadrado)
```

```python
L = 600 
r = 3
fig, ax = plt.subplots(figsize=(16,16))
ax.set_xlim(0, 600)
ax.set_ylim(0, 600)
ax.set_xticks([])
ax.set_yticks([])   
cuadrados(0, 0, L, r)
plt.show()
```

# b)

```python
def cuadra2(x1, y1, L, n):
    '''
        x1, y1: coordenadas del vértice superior izquierdo
        L: longitud del lado del cuadrado a subdividir
        n: número de iteraciones
    '''
    if n > 0:
        ra = np.random.randint(1, 5)

        if ra == 1:
            cuadra2(x1, y1, L/2, n-1)
            cuadra2(x1 + L/2, y1, L/2, n-1)
            cuadra2(x1 + L/2, y1 + L/2, L/2, n-1)
        elif ra == 2:
            cuadra2(x1, y1, L/2, n-1)
            cuadra2(x1 + L/2, y1, L/2, n-1)
            cuadra2(x1, y1 + L/2, L/2, n-1)
        elif ra == 3:
            cuadra2(x1, y1, L/2, n-1)
            cuadra2(x1, y1 + L/2, L/2, n-1)
            cuadra2(x1 + L/2, y1 + L/2, L/2, n-1)
        elif ra == 4:
            cuadra2(x1 + L/2, y1, L/2, n-1)
            cuadra2(x1, y1 + L/2, L/2, n-1)
            cuadra2(x1 + L/2, y1 + L/2, L/2, n-1)
    else:
        cuadrado = plt.Rectangle((x1, y1), L, L, color='r', fill=True)
        ax.add_patch(cuadrado)
```

```python
L = 600 
r = 8
fig, ax = plt.subplots(figsize=(16,16))
ax.set_xlim(0, 600)
ax.set_ylim(0, 600)
ax.set_xticks([])
ax.set_yticks([])   
cuadra2(0, 0, L, r)
plt.show()
```
