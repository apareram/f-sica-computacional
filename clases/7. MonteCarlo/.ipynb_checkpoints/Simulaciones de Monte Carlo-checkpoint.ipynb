# Simulaciones de Monte Carlo de un gas ideal

Supongamos $N$ partículas no interactuántes en una caja de lado $L$. La energía de una partícula viene dada por
$$E(n_x,n_y,n_z) = \frac{\pi^2\hbar^2}{2mL^2}(n_x^2+n_y^2+n_z^2)$$
y la energía del sistema completo es:
$$ E = \sum_{i=1}^N E(n_x^{(i)},n_y^{(i)},n_z^{(i)}).$$
En cada paso solamente se permitirá  que un átomo en particular se pueda mover  a un estado adyacente. 
Se escogen al azar  el átomo, el $n_i$ y si el cambio es $\pm 1$. En cda moviemiento la diferencia de energía del
sistema será:
$$\Delta E = \frac{\pi^2\hbar^2}{2mL^2}(\pm 2n_i + 1)$$
Datos: $L=1$, $k_B T = 10$, Unidades: $m=\hbar=1$

```python
from random import random,randrange,seed
import numpy as np
import matplotlib.pyplot as plt 

#seed(1000)
T = 10.0
N = 1
steps = 500000

# Crear un arreglo 2d para almacenar los números cuánticos
n = np.ones([N,3],int) #todas la partículas están inicialmente en el estado base
#n = np.random.randint(1,high = 3, size = (N,3))
# Main loop
eplot3 = []
E = 3*N*np.pi*np.pi/2
#E = (np.pi*np.pi/2)*np.sum(n**2)
for k in range(steps):

    # Escogencia de la partícula y del número cuántico a cambiar
    i = randrange(N)
    j = randrange(3)
    if random()<0.5:
        dn = 1
        dE = (2*n[i,j]+1)*np.pi*np.pi/2
    else:
        dn = -1
        dE = (-2*n[i,j]+1)*np.pi*np.pi/2

    # Metropolis
    if n[i,j]>1 or dn==1: #Permito que ocurra para todo n
        if dn == -1 or random()< np.exp(-dE/T):
            n[i,j] += dn
            E += dE

    eplot3.append(E)
```

# Simulación de una partícula en equilibrio con un baño térmico 
## (una dimensión)

La idea es simular un gas ideal clásico en una dimensión en contacto con un reservorio térmico a una temerpatura $T$. Como las partículas no interactúan basta simular una sola partícula. 

```python
from random import random,randrange,seed
import numpy as np
import matplotlib.pyplot as plt 

T = 10 #k_B T = 10
beta = 1./T
v0 = 10 #velocidad inicial
delta = 10 #maximo cambio de velocidades
steps = 1000000
k0 = 0.5*v0*v0
eplot = [k0]
Velocity = [v0]

for mcs in range(steps):
    vprueba = v0  + (2*random()-1)*delta
    kprueba = 0.5*vprueba*vprueba
    dk = kprueba-k0
    if dk < 0 or random() < np.exp(-beta*dk):
        v0 = vprueba
        k0 = kprueba
    eplot.append(k0)
    Velocity.append(v0)
```

```python
# Make the graph
plt.plot(eplot,'.')
plt.ylabel("Energy")
plt.show()
Eprom= np.mean(np.array(eplot[100:]))
print('<E> = ',Eprom,'kBT/2 = ',0.5*T,'error =',np.abs(Eprom-0.5*T)/0.5*T*100,'%')
```

### Media por intervalo

```python
eplot = np.array(eplot)
eprom = []
tiempo = []
intervalo = 100
t = 0
while True:
    eprom.append(eplot[t:t+intervalo].mean())
    t += intervalo
    tiempo.append(t)
    if t + intervalo > steps:
        break  
```

```python
plt.figure(figsize=(8,6))
plt.plot(tiempo,eprom)
plt.xlabel(r'$t$',fontsize = 14)
plt.ylabel(r'$\langle E \rangle (t)$',fontsize = 14)
plt.axhline(T/2,color = "red",dashes = (2,1),zorder=2)
plt.text(steps+1000,0.95*T/2,r'$\frac{1}{2}k_B T$',fontsize = 14)
plt.xlim(0,steps)
```

```python
plt.figure(figsize=(8,6))
n, bins, patches = plt.hist(Velocity, bins = 100,color = 'red',density = True)
x = np.linspace(-15,15,100)
Pb = np.sqrt(beta/(2*np.pi))*np.exp(-beta*x**2/2)
plt.plot(x,Pb,'k-')
plt.ylabel(r'$P(v)$',fontsize = 18)
plt.xlabel(r'$v$',fontsize = 18)
plt.text(2.5,0.12,r'$P_B(v) = \sqrt{\beta m/2\pi}\,\exp(-\beta m v^2/2)$')
```

# El modelo de Ising

Supongamos que tenemos $N$ partículas de spin $1/2$ dispuestos es una red. Supongamos que dichos spines interactúan vía un término de intercambio. La energía del sistema vendrá dada por:
$$ E = \sum_{i\ne j} J_{ij} s_is_j $$
Supongamos que solamente hay interacción entre primeros vecinos, la energía del sistema de $N$ spines será:
$$ E = J \sum_{\langle ij\rangle} s_is_j $$
Supongamos incialmente que nuestra red es cuadrada y que hay condiciones de borde periódicas en la frontera.

## Programa de Metropolis para el modelo de Ising

```python
# Consideraremos como variables gloables todo aquello que tenga que ver con
# la estructura del sistema que estamos modelando
# L: longitud de la red
# N: número de spines
# nn[s,j]: vecino j sel sitio s
# red[s]: areglo unidimensional que guarda en las posiciones de los spines del sistema
```

```python
# Estableciendo los vecinos de cada sito s
def vecino(s,j):
    global L,N
    if j == 0: #derecha
        if s%L ==L-1:
            return s-L+1
        else:
            return s+1
    elif j==1: #arriba
        if s//L == L-1:
            return s%L
        else:
            return s+L
    elif j==2: #izquierda
        if s%L == 0:
            return s+L-1
        else:
            return s-1
    elif j==3: #abajo
        if s//L == 0:
            return N-L+s
        else:
            return s-L
                
def boundaries():
    global N
    for s in range(N):
        for j in range(4):
            nn[s,j] = vecino(s,j)
```

```python
import numpy as np

global L,N

L=32
N=L*L
nn = np.zeros((N,4),dtype='uint32')
boundaries()
```

```python
#np.random.seed(100)

#Datos de la corrida
#temperature
# Tc = 2.269185
T = 1.269185
beta = 1/T

#red = 2*np.random.randint(0,2,N)-1 #Generación al azar de los spines
red = np.random.choice([1, -1],N) #Generación al azar de los spines
#red = np.ones(N) #Generación ordenada de los spines

#Calculo de la magnetización inicial
mag = red.sum()
Mag = [mag/N]

#Calculo de la energía inicial
J = 1   #termino de intercambio
ener = 0.
for s in range(N):
    for j in range(2):
        ener += red[s]*red[nn[s,j]]
ener = -J*ener
Energy = [ener/N]

#Arreglo que guarda las porbabilidades de Metropolis
expon = {4:np.exp(-J*beta*4),8:np.exp(-J*beta*8)}
```

```python
#Comienza Metroplis
Nmcs = 3000000

for mcs in range(Nmcs):
    sr = np.random.randint(N) #escogencia de  un sitio al azar
    dm = -2*red[sr] #posible cambio en mag
    de = 0
    for j in range(4):
        de += red[nn[sr,j]]
    de = -dm*de #posible cambio en ener (sin inclurir J)
    #Cambio de Metropolis
    if de <= 0 or np.random.random() < expon[de]:        
        ener += J*de
        mag += dm
        red[sr] = -red[sr]
    Mag.append(mag/N)
    Energy.append(ener/N)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(Mag,'b-',label = ("T=%4.2f" %T))
ax1.legend(loc = 'lower left', fontsize = 14)
ax1.set_xlabel(r'$mcs$',fontsize = 18)
ax1.set_ylabel(r'$M$',fontsize = 18)
ax1.set_title('Magnetización',fontsize = 24)

ax1 = fig.add_subplot(2, 1, 2)
ax1.plot(Energy,'b-',label = ("T=%4.2f" %T))
ax1.legend(loc = 'lower left', fontsize = 14)
ax1.set_xlabel(r'$mcs$',fontsize = 18)
ax1.set_ylabel(r'$E$',fontsize = 18)
ax1.set_title('Energía',fontsize = 24)

fig.tight_layout()
plt.show()
```

# Recocido simulado
## El problema del agente viajero

```python
from math import sqrt,exp
from numpy import empty
from random import random,randrange

from vpython import sphere,curve,rate,vector,mag,canvas
#from visual import sphere,curve,display,rate

N = 25
R = 0.02
Tmax = 10.0
Tmin = 1e-3
tau = 1e4

# Function to calculate the magnitude of a vector
#def mag(x):
#    return sqrt(x[0]**2+x[1]**2)

# Function to calculate the total length of the tour
def distance():
    s = 0.0
    for i in range(N):
        s += mag(r[i+1]-r[i])
    return s

# Choose N city locations and calculate the initial distance
r = []
for i in range(N):
    r.append(vector(random(),random(),0))
r.append(r[0])

D = distance()
Distancia = [D]
# Set up the graphics
canvas(center=vector(0.5,0.5,0))
for i in range(N):
    sphere(pos=r[i],radius=R)
l = curve(pos=r,radius=R/2)

# Main loop
t = 0
T = Tmax
while T>Tmin:

    # Cooling
    t += 1
    T = Tmax*exp(-t/tau)

    # Update the visualization every 100 moves
    if t%100==0:
        l.visible = False
        l = curve(pos=r,radius=R/2)
        rate(25)

    # Choose two cities to swap and make sure they are distinct
    i,j = randrange(1,N),randrange(1,N)
    while i==j:
        i,j = randrange(1,N),randrange(1,N)

    # Swap them and calculate the change in distance
    oldD = D
    dummy = r[i]
    r[i] = r[j]
    r[j] = dummy
    D = distance()
    deltaD = D - oldD

    # If the move is rejected, swap them back again
    if random()>exp(-deltaD/T):
        dummy = r[i]
        r[i] = r[j]
        r[j] = dummy
        D = oldD
    Distancia.append(D)
```

```python
import matplotlib.pyplot as plt
plt.plot(Distancia,'b-')
plt.ylabel(r'$D$')
plt.xlabel(r'$pasos$')
```
